```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris = datasets.load_iris(as_frame=True)
```

```python
iris["frame"]
```

```python
# Visual representation of the data. We can see that there is no NULL value in our data
import missingno as msno
msno.matrix(iris["frame"])
```

```python
print(list(iris))
```

```python
print(list(iris.data))
```

```python
#print(iris.DESCR)
```

```python
iris2 = iris.data[['sepal length (cm)', 'petal length (cm)']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris2,iris.target)
```

```python
from sklearn.svm import SVC
clf1 = SVC(kernel='linear')
clf2 = SVC(kernel="rbf")
clf3 = SVC(kernel="poly", degree=3)
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay  
cm1 = confusion_matrix(y_test, clf1.predict(X_test), labels=clf1.classes_)
cm2 = confusion_matrix(y_test, clf2.predict(X_test), labels=clf1.classes_) 
cm3 = confusion_matrix(y_test, clf3.predict(X_test), labels=clf1.classes_) 

disp = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=clf1.classes_)
disp.plot()
```

```python
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=clf1.classes_)
disp2.plot()
```

```python
disp3 = ConfusionMatrixDisplay(confusion_matrix=cm3, display_labels=clf1.classes_)
disp3.plot()
```

```python
from matplotlib.colors import ListedColormap
x_length = np.linspace(0,40,100)
y_length = np.linspace(0,40,100)
xx, yy = np.meshgrid(x_length,y_length)


X = iris2.values[:,0:2]
y = iris.target
    
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

x_space = np.linspace(x_min, x_max,100)
y_space = np.linspace(y_min, y_max,100)
xx, yy = np.meshgrid(x_space,y_space)
XXYY = np.array([xx.ravel(), yy.ravel()]).T   
y_pred = clf1.predict(XXYY)
y_pred = y_pred.reshape(xx.shape)  

plt.contourf(xx, yy, y_pred, alpha = 0.75, cmap = ListedColormap(('red', 'green',"blue")))               
plt.scatter(X[:,0],X[:,1], c = y, edgecolor = "black")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

Problem 3

```python
digits = datasets.load_digits(as_frame=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
from sklearn.neural_network import MLPClassifier
clf1 = MLPClassifier(solver='adam', hidden_layer_sizes=(1,200), random_state=1,max_iter=500)
clf2 = MLPClassifier(solver='adam', hidden_layer_sizes=(2,500), random_state=1,max_iter=500)
clf3 = MLPClassifier(solver='sgd', hidden_layer_sizes=(52,500), random_state=1,max_iter=500)
```

```python
clf1.fit(X_train, y_train)
```

```python
clf2.fit(X_train, y_train)
```

```python
clf3.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
acc1 = accuracy_score(y_test, clf1.predict(X_test))
pre1 = precision_score(y_test, clf1.predict(X_test), average='micro',zero_division=1)
rec1 = recall_score(y_test, clf1.predict(X_test), average='macro', zero_division=1)
print(f"Accuracy: {acc1}, Precision: {pre1}, Recall: {rec1}")
```

```python
acc2 = accuracy_score(y_test, clf2.predict(X_test))
pre2 = precision_score(y_test, clf2.predict(X_test), average='macro',zero_division=1)
rec2 = recall_score(y_test, clf2.predict(X_test), average='macro',zero_division=1)
print(f"Accuracy: {acc2}, Precision: {pre2}, Recall: {rec2}")
```

```python
acc3 = accuracy_score(y_test, clf3.predict(X_test))
pre3 = precision_score(y_test, clf3.predict(X_test), average='macro',zero_division=1)
rec3 = recall_score(y_test, clf3.predict(X_test), average='macro',zero_division=1)
print(f"Accuracy: {acc3}, Precision: {pre3}, Recall: {rec3}")
```

```python
cm1 = confusion_matrix(y_test, clf1.predict(X_test), labels=clf1.classes_)
cm2 = confusion_matrix(y_test, clf2.predict(X_test), labels=clf1.classes_)
cm3 = confusion_matrix(y_test, clf3.predict(X_test), labels=clf1.classes_) 

disp = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=clf1.classes_)
disp.plot()
```
